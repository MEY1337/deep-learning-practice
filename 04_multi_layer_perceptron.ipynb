# 코드를 통해서 XOR 문제 해결하기

```python
# 가중치와 바이어스
w11 = np.array([-2, -2])
w12 = np.array([2, 2])
w2 = np.array([1, 1])

b1 = 3
b2 = -1
b3 = -1
```

```python
# 퍼셉트론
def MLP(x, w, b):
    y = np.sum(w * x) + b

    if y <= 0:
        return 0
    else:
        return 1
```

```python
# NAND 게이트
def NAND(x1,x2):
    return MLP(np.array([x1, x2]), w11, b1)

# OR 게이트
def OR(x1,x2):
    return MLP(np.array([x1, x2]), w12, b2)

# AND 게이트
def AND(x1,x2):
    return MLP(np.array([x1, x2]), w2, b3)

# XOR 게이트
def XOR(x1,x2):
    return AND(NAND(x1, x2),OR(x1,x2))
```

```python
# x1, x2 값을 번갈아 대입해 가며 최종 값 출력
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = XOR(x[0], x[1])
    print("입력 값: " + str(x) + " 출력 값: " + str(y))
```

# Multi Perceptron Class 생성을 통해 해결하기

```python
# XOR 문제를 해결하기 위한 다층 퍼셉트론
class SimpleMLP:
    def __init__(self):
        # 수동으로 설정한 가중치와 바이어스
        self.weights_input_hidden = np.array([[-2, 2], [-2, 2]])  # x1, x2 -> n1, n2 계산을 위한 가중치 배열
        self.bias_hidden = np.array([3, -1])  # x1, x2 -> n1, n2

        # 출력층: 1개의 뉴런
        self.weights_hidden_output = np.array([1, 1])  # n1, n2 -> y 가중치
        self.bias_output = -1  # n1, n2 -> y 바이어스

    def forward(self, X):
        # 은닉층 계산
        hidden_input = np.dot(X, self.weights_input_hidden) + self.bias_hidden
        hidden_output = np.where(hidden_input <= 0, 0, 1)
        print(f"hidden_output : {hidden_input}")

        # 출력층 계산
        final_input = np.dot(hidden_output, self.weights_hidden_output) + self.bias_output
        final_output = np.where(final_input <= 0, 0, 1)
        print(f"final_output : {final_input}")

        return final_output
```

```python
# XOR 데이터셋 정의
XOR_X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # 입력값
XOR_y = np.array([0, 1, 1, 0])  # XOR 출력값
```

```python
# 모델 초기화
mlp = SimpleMLP()
```

```python
# 예측
predictions = mlp.forward(XOR_X)
# 결과 출력
print(f"XOR 문제 예측 결과: {predictions}")
```

# Perceptron을 이용한 NAND Gate와 OR Gate 구현

퍼셉트론의 출력식은 다음과 같다.

$$
y =
\begin{cases}
1, & w_1 x_1 + w_2 x_2 + b > 0 \
0, & \text{otherwise}
\end{cases}
$$

또한 내부 계산값은 다음과 같이 정의된다.

$$
z = w_1 x_1 + w_2 x_2 + b
$$

퍼셉트론의 입력은 다음과 같은 열 벡터 형태로 표현할 수 있다.

$$
\mathbf{x} =
\begin{bmatrix}
x_1 \
x_2
\end{bmatrix}
$$

가중치 또한 열 벡터로 표현된다.

$$
\mathbf{w} =
\begin{bmatrix}
w_1 \
w_2
\end{bmatrix}
$$

따라서 퍼셉트론의 연산은 벡터의 내적 형태로 다음과 같이 나타낼 수 있다.

$$
z = \mathbf{w}^T \mathbf{x} + b
$$

이때 출력값을 결정하는 함수는 활성화 함수이며, 퍼셉트론에서는 계단 함수(step function)를 사용한다.

$$
\phi(z) =
\begin{cases}
1, & z > 0 \
0, & z \leq 0
\end{cases}
$$

## 1. NAND Gate

NAND Gate의 진리표는 다음과 같다.

$$
(0,0) \rightarrow 1,\quad
(0,1) \rightarrow 1,\quad
(1,0) \rightarrow 1,\quad
(1,1) \rightarrow 0
$$

NAND Gate의 가중치는 열 벡터 형태로 다음과 같이 표현할 수 있다.

$$
\mathbf{w} =
\begin{bmatrix}
-2 \
-2
\end{bmatrix}
$$

따라서 다음과 같은 부등식을 만족해야 한다.

$$
b > 0
$$

$$
-2x_2 + b > 0
$$

$$
-2x_1 + b > 0
$$

$$
-2x_1 - 2x_2 + b \leq 0
$$

위 조건을 만족하는 하나의 해는 다음과 같다.

$$
b = 3
$$

이를 검증하면 다음과 같다.

$$
(0,0):\quad 0 + 0 + 3 = 3 > 0
$$

$$
(0,1):\quad 0 - 2 + 3 = 1 > 0
$$

$$
(1,0):\quad -2 + 0 + 3 = 1 > 0
$$

$$
(1,1):\quad -2 - 2 + 3 = -1 \leq 0
$$

따라서 위 가중치는 NAND Gate를 만족한다.

## 2. OR Gate

OR Gate의 진리표는 다음과 같다.

$$
(0,0) \rightarrow 0,\quad
(0,1) \rightarrow 1,\quad
(1,0) \rightarrow 1,\quad
(1,1) \rightarrow 1
$$

OR Gate의 가중치는 열 벡터 형태로 다음과 같이 표현할 수 있다.

$$
\mathbf{w} =
\begin{bmatrix}
2 \
2
\end{bmatrix}
$$

따라서 다음과 같은 부등식을 만족해야 한다.

$$
b \leq 0
$$

$$
2x_2 + b > 0
$$

$$
2x_1 + b > 0
$$

$$
2x_1 + 2x_2 + b > 0
$$

위 조건을 만족하는 하나의 해는 다음과 같다.

$$
b = -1
$$

이를 검증하면 다음과 같다.

$$
(0,0):\quad 0 + 0 - 1 = -1 \leq 0
$$

$$
(0,1):\quad 0 + 2 - 1 = 1 > 0
$$

$$
(1,0):\quad 2 + 0 - 1 = 1 > 0
$$

$$
(1,1):\quad 2 + 2 - 1 = 3 > 0
$$

따라서 위 가중치는 OR Gate를 만족한다.

## 3. 다층 퍼셉트론

다층 퍼셉트론은 입력층, 은닉층, 출력층으로 구성된 신경망 구조이다. 단층 퍼셉트론과 달리 여러 층을 통해 복잡한 패턴을 학습할 수 있으며, 각 층에서 활성화 함수를 사용하여 비선형성을 추가한다. 이러한 구조를 통해 딥러닝은 단순한 선형 분류를 넘어 복잡한 문제를 해결할 수 있다.

## 3.1 딥러닝에서 중요한 요소

딥러닝에서 중요한 요소 중 하나는 활성화 함수이다. 활성화 함수는 뉴런의 출력값을 결정하며, 비선형성을 추가하여 모델이 복잡한 패턴을 학습할 수 있도록 한다. 만약 활성화 함수가 없다면 모델은 단순한 선형 변환만 수행하게 되어 표현력이 제한된다.

## 3.2 딥러닝과 기존 머신러닝의 차이

딥러닝은 기존 머신러닝과 달리 특징을 사람이 직접 추출하지 않고, 신경망이 자동으로 학습한다는 점에서 차이가 있다. 또한 활성화 함수를 활용하여 비선형적인 문제를 효과적으로 해결할 수 있다.
